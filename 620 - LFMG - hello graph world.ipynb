For this assignment I wanted to change the nodes to something that spoke more to me: food. 
So, we will start by naming the graph, then respectively the nodes and how the edges will connect the nodes.
I particularly enjoyed choosing color names that would maintain the overall theme.

```python
import networkx as nx
import matplotlib.pyplot as plt

# to create an undirected graph
Hungry_Kite = nx.Graph()

# Defining Yummy nodes 
nodes = ["Pizza", "Tacos", "Sushi", "Pasta", "Burger", 
    "Salad", "Ramen", "Steak", "Curry", "Falafel"]
Hungry_Kite.add_nodes_from(nodes)

# Define edges to connect dishes
edges = [
    ("Pizza", "Tacos"), ("Pizza", "Sushi"), ("Pizza", "Pasta"),
    ("Tacos", "Pasta"), ("Tacos", "Burger"),
    ("Sushi", "Salad"), ("Sushi", "Pasta"),
    ("Pasta", "Salad"), ("Pasta", "Ramen"),
    ("Burger", "Ramen"),
    ("Salad", "Steak"),
    ("Ramen", "Steak"),
    ("Steak", "Curry"),
    ("Curry", "Falafel")
]
Hungry_Kite.add_edges_from(edges)

# plotting 
plt.figure(figsize=(7,6))
pos = nx.spring_layout(Hungry_Kite, seed=5)
nx.draw(Hungry_Kite, pos, with_labels=True, node_color='papayawhip', edge_color='darksalmon', node_size=1400, font_size=10)
plt.title("Yummy Kite Graph")
plt.show()
```
